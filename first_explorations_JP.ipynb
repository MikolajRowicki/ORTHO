```python
import pandas as pd
import glob
import os
# Pobranie pierwszych 40 plików CSV w katalogu data/full_dta
csv_files = sorted(glob.glob("data/full_data/*.csv"))[:40]

# Inicjalizacja ramki danych
limited_data = pd.DataFrame()

# Wczytanie pierwszych 40 plików
for file in csv_files:
    try:
        df = pd.read_csv(file)
        
        # Dodanie danych do ramki
        limited_data = pd.concat([limited_data, df], ignore_index=True)
    except Exception as e:
        print(f"Błąd podczas przetwarzania pliku {file}: {e}")

# Wyświetlenie kilku pierwszych wierszy
print('--- Pierwsze 40 plików ---')
print(limited_data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytanie danych (dostosuj ścieżkę i separator)
df = pd.read_csv('sciezka/do/pliku.csv', sep=',')

# --- KROK 1 i 2: Grupowanie po ID ---
grouped = df.groupby('ID')

# --- KROK 3: Filtrujemy całe grupy, jeśli w dowolnym wierszu jest Completed.1 == True i Track_ID == 3 ---
def gra_spelnia_warunek(grupa):
    # Szukamy w grupie wiersza z Completed.1 == True i Track_ID == 3
    mask = (grupa['Completed.1'] == True) & (grupa['Track_ID'] == 3)
    return mask.any()

df_filtered = grouped.filter(gra_spelnia_warunek)

# W tym momencie df_filtered zawiera PEŁNE grupy (wszystkie wiersze dla danej gry),
# jeśli ta gra ma przynajmniej jeden wiersz z Completed.1 == True i Track_ID == 3.

# --- KROK 4: Rysujemy trajektorie na podstawie wierszy, które mają Points_X i Points_Y ---
grouped_filtered = df_filtered.groupby('ID')

plt.figure(figsize=(10, 6))

for game_id, game_data in grouped_filtered:
    # Wyodrębniamy tylko te wiersze z danej grupy, które mają faktyczne współrzędne (nie NaN)
    game_data_points = game_data.dropna(subset=['Points_X', 'Points_Y'])

    # Opcjonalnie można posortować punkty wg czasu, jeśli kolumna Points_Time jest do tego odpowiednia:
    # game_data_points = game_data_points.sort_values('Points_Time')

    if not game_data_points.empty:
        x = game_data_points['Points_X'].values
        y = game_data_points['Points_Y'].values
        plt.plot(x, y, marker='o', label=f'Gra {game_id}')

plt.xlabel('Points_X')
plt.ylabel('Points_Y')
plt.title('Trajektorie gier (Completed.1=True, Track_ID=3)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Przyjmujemy, że Twoja ramka danych to:
df = pd.read_csv("./data/full_data/games.csv")  # lub: df = pd.read_csv("nazwa_pliku.csv")

# Ustawienia wykresów
sns.set(style="whitegrid")
plt.rcParams.update({'figure.max_open_warning': 0})

#########################################
# 1. Histogram liczby uruchomień gry przez zespoły
#    (ile zespołów uruchamiały grę k razy)
#########################################
# Liczymy, ile razy każdy zespół uruchomił grę:
team_counts = df["Team_Name"].value_counts(dropna=True)
# Grupujemy te wartości, aby dowiedzieć się, ile zespołów miało k uruchomień:
freq_counts = team_counts.value_counts().sort_index()

# Filtrujemy tylko te kategorie, gdzie liczba zespołów (y) jest > 10:
freq_counts_filtered = freq_counts[freq_counts > 10]

plt.figure(figsize=(10,6))
ax = sns.barplot(x=freq_counts_filtered.index, y=freq_counts_filtered.values, palette="Blues_d")
plt.title("Histogram: Liczba zespołów, które uruchamiały grę k razy (y > 10)")
plt.xlabel("Liczba uruchomień gry (k)")
plt.ylabel("Liczba zespołów")
# Poprawa widoczności osi x:
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

#########################################
# 2. Histogramy związane z wiekiem (Age_X_axis, Age_Y_axis)
#    dla każdej drużyny licz wiek jednokrotnie
#########################################
# Usuwamy duplikaty według drużyny, aby każda była liczona raz:
df_unique = df.drop_duplicates(subset=["Team_Name"])

# Konwersja kolumn na wartości liczbowe (w razie, gdyby były zapisane jako stringi)
age_x_unique = pd.to_numeric(df_unique["Age_X_axis"], errors='coerce')
age_y_unique = pd.to_numeric(df_unique["Age_Y_axis"], errors='coerce')

# Połączone dane – każdy wiek z unikalnej drużyny
combined_ages_unique = pd.concat([age_x_unique.dropna(), age_y_unique.dropna()])

# Obliczenie bezwzględnej różnicy wieku (dla unikalnych drużyn)
abs_diff_unique = (age_x_unique - age_y_unique).abs().dropna()

# Tworzymy dwa wykresy obok siebie:
fig, axs = plt.subplots(1, 2, figsize=(14,6))

# Histogram połączonych wartości wieku
sns.histplot(combined_ages_unique, bins=20, kde=True, ax=axs[0], color="skyblue")
axs[0].set_title("Histogram wieku (Age_X i Age_Y, jednokrotnie dla drużyny)")
axs[0].set_xlabel("Wartość wieku")
axs[0].set_ylabel("Liczba drużyn")

# Histogram bezwzględnej różnicy wieku
sns.histplot(abs_diff_unique, bins=20, kde=True, ax=axs[1], color="salmon")
axs[1].set_title("Histogram |Age_X - Age_Y| (jednokrotnie dla drużyny)")
axs[1].set_xlabel("Wartość różnicy")
axs[1].set_ylabel("Liczba drużyn")

plt.tight_layout()
plt.show()

#########################################
# 3. Rozkład poziomów trudności (Difficulty_Level)
#    - Pie chart
#    - Bar chart z rosnącym porządkiem na osi x
#########################################
# Sortujemy według rosnącej wartości poziomów trudności:
difficulty_counts = df["Difficulty_Level"].value_counts(dropna=False).sort_index()

# Pie chart
plt.figure(figsize=(8,8))
plt.pie(difficulty_counts.values, labels=difficulty_counts.index.astype(str), autopct='%1.1f%%', 
        colors=sns.color_palette("pastel"))
plt.title("Rozkład poziomów trudności (Difficulty_Level) - Pie Chart")
plt.tight_layout()
plt.show()

# Bar chart
plt.figure(figsize=(8,6))
sns.barplot(x=difficulty_counts.index.astype(str), y=difficulty_counts.values, palette="pastel")
plt.title("Rozkład poziomów trudności (Difficulty_Level) - Bar Chart")
plt.xlabel("Difficulty_Level")
plt.ylabel("Liczba uruchomień")
plt.tight_layout()
plt.show()


#########################################
# 4. Wykres 5: Estetyczny wykres liniowy – liczba uruchomień gry według daty
#########################################
# Liczymy liczbę uruchomień na każdą datę i sortujemy wg daty
df["Date"] = pd.to_datetime(df["Date"].astype(str).str[:8], format="%Y%m%d", errors='coerce')

# Grupujemy dane – liczymy liczbę uruchomień gry dla każdej daty
date_counts = df["Date"].value_counts().sort_index()

# Tworzymy pełny zakres dat dla 2024 roku (od 1 stycznia do 31 grudnia)
full_dates = pd.date_range(start="2024-01-01", end="2024-12-31", freq="D")

# Uzupełniamy braki – jeśli w oryginalnych danych nie ma wpisu dla danej daty, przypisujemy 0
date_counts = date_counts.reindex(full_dates, fill_value=0)
date_counts.index.name = "Date"

# Przekształcamy serię w DataFrame, aby wygodnie przekazać dane do Plotly Express
date_counts_df = date_counts.reset_index(name="Liczba uruchomień")

# Generujemy interaktywny wykres liniowy
fig = px.line(date_counts_df, x="Date", y="Liczba uruchomień", markers=True,
              title="Liczba uruchomień gry według daty (2024)")
# Dodatkowe ustawienia osi x – etykiety co miesiąc, format dd-mm-YYYY
fig.update_layout(
    xaxis_title="Data",
    yaxis_title="Liczba uruchomień",
    xaxis=dict(
        dtick="M1",
        tickformat="%d-%m-%Y"
    ),
    hovermode="x unified"
)
fig.show()

#########################################
# 5. Wykres: Liczba osób, które wybrały poszczególne języki gry
#########################################
language_counts = df["Selected_Language"].value_counts(dropna=True)

plt.figure(figsize=(8,6))
sns.barplot(x=language_counts.index.astype(str), y=language_counts.values, palette="viridis")
plt.title("Liczba osób wybierających poszczególne języki gry")
plt.xlabel("Wybrany język")
plt.ylabel("Liczba osób")
plt.tight_layout()
plt.show()

#########################################
# 6. Wykres: Numer toru (Track_ID) w podziale na Completed.1 (True/False)
#    oś x: Track_ID, oś y: liczba uruchomień; wykres grupowany według wartości w Completed.1
#########################################
# Grupujemy dane wg 'Track_ID' i 'Completed.1'
track_completed_counts = df.groupby(["Track_ID", "Track_Completed"]).size().unstack(fill_value=0)

plt.figure(figsize=(10,6))
track_completed_counts.plot(kind='bar', figsize=(10,6))
plt.title("Liczba uruchomień wg toru i Completed.1")
plt.xlabel("Track_ID")
plt.ylabel("Liczba uruchomień")
plt.xticks(rotation=45)
plt.legend(title="Completed.1", labels=["False", "True"])
plt.tight_layout()
plt.show()
```
